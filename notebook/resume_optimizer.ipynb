```python
# ============================================
# AI Resume Rewrite Optimizer (Advanced)
# Self-Refinement + Before/After Comparison
# ============================================

import os
import re
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_community.document_loaders import PyPDFLoader

# -------------------------
# Load API Key
# -------------------------
load_dotenv()
GOOGLE_API_KEY = os.getenv("GEMINI_API_KEY")

# -------------------------
# Initialize LLM
# -------------------------
llm = ChatGoogleGenerativeAI(
    model="models/gemma-3-27b-it",
    temperature=0.4
)

# -------------------------
# Resume Loader
# -------------------------
def load_resume_text(pdf_path: str) -> str:
    loader = PyPDFLoader(pdf_path)
    pages = loader.load()
    return "\n".join([page.page_content for page in pages])


# -------------------------
# Prompt 1 — Initial Rewrite
# -------------------------
rewrite_prompt = PromptTemplate(
    template="""
You are a senior executive resume writer.

Rewrite and optimize the resume below.

Rules:
- Do NOT fabricate new roles or achievements.
- Strengthen action verbs.
- Add measurable impact where context allows.
- Optimize for ATS keywords naturally.
- Improve clarity and conciseness.
- Keep original section structure.

Resume:
{resume_text}

Return only the rewritten resume.
""",
    input_variables=["resume_text"]
)

# -------------------------
# Prompt 2 — Critique Rewritten Resume
# -------------------------
critique_prompt = PromptTemplate(
    template="""
You are a resume quality auditor.

Analyze the rewritten resume below and identify:
- Remaining weaknesses
- Areas lacking quantification
- Overly generic statements
- Opportunities to increase impact

Provide improvement suggestions clearly.

Rewritten Resume:
{rewritten_resume}
""",
    input_variables=["rewritten_resume"]
)

# -------------------------
# Prompt 3 — Refinement Pass
# -------------------------
refine_prompt = PromptTemplate(
    template="""
You are a senior resume optimizer.

Improve the rewritten resume further based on the critique below.

Critique:
{critique}

Resume:
{rewritten_resume}

Rules:
- Increase impact.
- Strengthen quantification.
- Maintain factual integrity.
- Do NOT fabricate numbers if not inferable.
- Keep structure intact.

Return the final optimized resume only.
""",
    input_variables=["critique", "rewritten_resume"]
)

# -------------------------
# Quantification Counter
# -------------------------
def count_quantified_statements(text):
    pattern = r"\b\d+%?|\$\d+|\d+\+?"
    matches = re.findall(pattern, text)
    return len(matches)

# -------------------------
# Main Optimization Pipeline
# -------------------------
def optimize_resume(pdf_path: str):
    
    original_resume = load_resume_text(pdf_path)

    # ---- Pass 1: Rewrite ----
    rewrite_response = llm.invoke(
        rewrite_prompt.format(resume_text=original_resume)
    )
    rewritten_resume = rewrite_response.content

    # ---- Pass 2: Critique ----
    critique_response = llm.invoke(
        critique_prompt.format(rewritten_resume=rewritten_resume)
    )
    critique = critique_response.content

    # ---- Pass 3: Refinement ----
    final_response = llm.invoke(
        refine_prompt.format(
            critique=critique,
            rewritten_resume=rewritten_resume
        )
    )
    final_resume = final_response.content

    # ---- Metrics ----
    before_quant = count_quantified_statements(original_resume)
    after_quant = count_quantified_statements(final_resume)

    improvement_summary = {
        "quantified_statements_before": before_quant,
        "quantified_statements_after": after_quant,
        "improvement_delta": after_quant - before_quant
    }

    return {
        "original_resume": original_resume,
        "final_resume": final_resume,
        "improvement_summary": improvement_summary
    }


# -------------------------
# RUN
# -------------------------
if __name__ == "__main__":
    result = optimize_resume(r"C:\Python Data Analysis\LLM\AI Resume Critique\cv_computer-vision.pdf")
    
    print("========== IMPROVEMENT SUMMARY ==========")
    print(result["improvement_summary"])
    
    print("\n========== FINAL OPTIMIZED RESUME ==========\n")
    print(result["final_resume"])
```
